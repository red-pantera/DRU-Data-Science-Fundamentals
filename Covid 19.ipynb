```python
import pandas as pd
```

```python
covid_df = pd.read_csv('covid-data.csv', parse_dates=['date'])
```

```python
covid_df
```

```python
covid_ua = covid_df[covid_df.location == 'Ukraine']
covid_ua
```

```python
covid_ua.columns
```

```python
covid_ua[['total_vaccinations', 'people_vaccinated', 'people_fully_vaccinated', 'new_vaccinations']]
```

```python
covid_ua.drop(['iso_code', 'continent', 'location', 'handwashing_facilities'], axis = 1)
```

```python
covid_ua_stat = covid_ua[['date', 'new_cases', 'new_deaths', 'new_cases_per_million', 'new_tests', 'new_vaccinations']]
covid_ua_stat
```

```python
total_cases = covid_ua_stat.new_cases.sum()
total_deaths = covid_ua_stat.new_deaths.sum()

print('The number of reported cases is {} and the number of reported deaths is {}.'.format(int(total_cases), int(total_deaths)))
```

```python
death_rate = total_deaths / total_cases 

print("The overall reported death rate in Ukraine is {:.2f} %.".format(death_rate*100))
```

```python
total_tests = covid_ua_stat.new_tests.sum()
```

```python
positive_rate = total_cases / total_tests
```

```python
print('{:.2f}% of tests in Ukraine led to a positive diagnosis.'.format(positive_rate*100))
```
